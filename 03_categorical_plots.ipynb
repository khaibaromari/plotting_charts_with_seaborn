```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the data frames
mpg = sns.load_dataset("mpg")

# set the themes
sns.set_theme(style="whitegrid")
```

```python
# plotting a simple Count Plot
sns.catplot(x='cylinders', data=mpg, kind='count')
plt.show()
```

```python
# rotated count plot
sns.catplot(y='cylinders', data=mpg, kind='count')
plt.show()
```

```python
# ordering the bars
category_order=[8,6,5,4,3]
sns.catplot(x='cylinders', data=mpg, kind='count', order=category_order)
plt.show()
```

```python
# Divided the plot into subplots
sns.catplot(y='cylinders', data=mpg, col='origin', kind='count')
```

```python
# Bar plot average of mpg by number of cylinders
# it's advised to map the categorical variable to the x-axis
sns.catplot(x="cylinders", y="mpg", data=mpg, kind="bar")
# sns.catplot(x="cylinders", y="mpg", data=mpg, kind="bar", errorbar=None)
```

```python
# plot the bars horizontally
sns.catplot(x="mpg", y="origin", data=mpg, kind="bar")
plt.show()
```

```python
# point plot
# point plot is used to plot a categorical variable in x axis and mean of a continuous var by y axis
sns.catplot(x='cylinders', y='mpg', data=mpg, kind='point', hue='origin', alpha=.4)
```

```python
# point plot with removed joins
sns.catplot(x='cylinders', y='mpg', data=mpg, kind='point', linestyle='none', estimator='median', capsize=0.2)
```
