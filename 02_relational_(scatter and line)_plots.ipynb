```python
# adding more features to the plot:
# size: the size feature works best if the mapped variable is type of categorical
sns.relplot(x='horsepower', y='mpg', data=mpg, size='cylinders', hue='cylinders', kind='scatter')
plt.show()
```

```python
# style (shape):
sns.relplot(x='acceleration', y='mpg', data=mpg, style='origin',kind='scatter', hue='origin')
plt.show()
```

```python
# alpha (transparency)
sns.relplot(x='acceleration', y='mpg', data=mpg, kind='scatter', hue='origin', alpha=0.6)
plt.show()
```

```python
# Line chart

# plot a simple line chart
sns.relplot(x='model_year', y='mpg', data=mpg, kind='line', errorbar=None)
plt.show()
```

```python
# introducing errorbar to the plot as mean (if there are multiple y for the same x)
sns.relplot(x='model_year', y='mpg', data=mpg, kind='line')
plt.show()
```

```python
# introducing errorbar in form of SD (Standard Deviation) to the plot
sns.relplot(x='model_year', y='mpg', data=mpg, kind='line', errorbar='sd')
plt.show()
```

```python
# dividing the lines based on a variable
sns.relplot(x="model_year", y="horsepower", data=mpg, kind="line", errorbar=None, markers=True, dashes=False, style='origin')
plt.show()
```
