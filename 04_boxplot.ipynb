```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the data frames
mpg = sns.load_dataset("mpg")
```

```python
# plotting a smiple box plot
sns.catplot(x='origin', y='horsepower', data=mpg, kind='box')
```

```python
# ordered boxplot
sns.catplot(x='origin', y='horsepower', data=mpg, kind='box', order=['japan', 'europe', 'usa'])
```

```python
# introduce the third variable as color
sns.catplot(x="origin", y="horsepower",
            data=mpg,
            kind="box",
            hue="cylinders")
```

```python
# to extend the whiskers
sns.catplot(x='origin', y='horsepower', data=mpg, kind='box', whis=2)
```

```python
# to extend the upper and lower whiskers to percentiles
sns.catplot(x='origin', y='horsepower', data=mpg, kind='box', whis=[5, 95])
```
